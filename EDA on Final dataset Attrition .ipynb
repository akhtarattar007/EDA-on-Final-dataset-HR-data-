## 1. Travelling  detail of employee

```python
# Travelling  detail of employee

Travel_details = df['BusinessTravel'].value_counts().reset_index()
Travel_details 
```

```python
fig, ax = plt.subplots(figsize=(10,6))
Travel_barplot= sns.barplot(data=Travel_details,x='index',y='BusinessTravel')
Travel_barplot.set(xlabel ="Type of Business Travel by employee", ylabel = "Counts", title ="Barplot for Business Travel")
plt.show()
```

### Conslusion : Most of the employee are travelling rarely.

##2. Working mode of employee

```python
# WOrk mode of employee

work_mode =pd.DataFrame(df['Mode_of_work']).value_counts().reset_index()
work_mode
```

```python
work_mode.rename(columns = {0:'No of Employee'}, inplace = True)
work_mode
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=work_mode, x='Mode_of_work', y='No of Employee')
```

### Conclusion : Company have mostly equal count of employee that works WFH: Work From Home and WFO: Work From office

##3.Reason why employee are leaving company

```python
df.head(2)
```

```python
# why employee are leaving company
leaving = pd.DataFrame(df['Status_of_leaving'].value_counts())
leaving
```

```python
leaving.rename(columns = {'Status_of_leaving':'Counts'},inplace=True)
leaving
```

```python
explode_val = (0.1,0,0,0,0)
```

```python
plt.pie(leaving,labels =leaving.index,autopct='%1.1f%%',explode=explode_val)
```

### Conclusion : More than 40% of employee are leaving company because lo Dept. Head and low salary .

##5. Expenditure on Salary of employee according to job role.

```python
# Monthly salary of Male and Female Employee according to Job role.
# Ploting using seaborn(sns) countplot
fig = plt.figure(figsize=(10,5))
sns.countplot(data=df['MonthlyIncome'], x = df['JobRole'],hue=df["Gender"])
plt.xlabel('JobRole', fontsize=12)
plt.xticks(rotation=90,fontsize=12);
```
